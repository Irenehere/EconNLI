```python
import wikipedia
from tqdm import tqdm
import os

wikipedia.set_lang('en')

page_list_root = ['Microeconomics', 'Macroeconomics','Finance']

page_rank_1 = []

for page in page_list_root:
    page = wikipedia.page(page)
    page_rank_1+=page.links
page_rank_1 = list(set(page_rank_1))

for page in tqdm(page_rank_1):
    if os.path.exists('../Corpus/Wiki_Corpus/'+page+'.txt'):
        continue
    try:
        page = wikipedia.page(page,auto_suggest=False)
    except:
        continue
    
    with open('../Corpus/Wiki_Corpus/'+page.title+'.txt', 'w', encoding='utf-8') as f:
        f.write(page.content)
```
